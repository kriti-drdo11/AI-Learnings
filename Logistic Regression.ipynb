##### Fature Scaling

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(x)
```

```python
x_scaled
```

```python
pd.DataFrame(x_scaled).describe()
```

##### Cross Validation

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test , y_train,y_test = train_test_split(x_scaled,y,random_state=0)
```

```python
x_train
```

```python
y_train
```

##### Build the Model

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression()
```

```python
#train Algorithm
classifier.fit(x_train,y_train)
```

##### predict on unknown data

```python
y_pred = classifier.predict(x_test)
```

```python
y_pred
```

```python
y_test.values
```

```python
result = pd.DataFrame({
    'actual': y_test,
    'predicted': y_pred
})
result
```

```python
y_test.value_counts()
```

##### Performance Evaluation

```python
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import accuracy_score, ConfusionMatrixDisplay
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
cm = confusion_matrix(y_test,y_pred)
```

```python
cm.diagonal().sum() / cm.sum()
```

##### Print report

```python
print(classification_report(y_test,y_pred))
```

```python
#accuracy of class 0/ Recall -0/senstivity of Algorithm
r0 = 67/68
r0
```

```python
# Accuracy of class 1/recall -1/Specificity of Algorithm
r1 = 22/32
r1
```

```python
# precision of class 0/ Positive Predictive Value
p0 = 67/77
p0
```

```python
# precision of class 1/ negative Predictive Value
p1 = 22/23
p1
```

```python
# f1-score -0 
2 * p0 * r0/(p0 + r0)
```

```python
# f1-score -1 
2 * p1 * r1/(p1 + r1)
```
