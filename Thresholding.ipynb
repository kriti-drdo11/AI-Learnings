```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('images/lenna.png',0)
```

##### BINARY & Inverted

```python
new_img = img > 127
inv_img = img < 127
```

```python
new_img
```

print(new_img)

```python

plt.figure(figsize=(16,9))
plt.subplot(131)
plt.title('original')
plt.imshow(img, cmap = 'gray')
plt.subplot(132)
plt.title('Binary')
plt.imshow(new_img, cmap = 'gray')
plt.subplot(133)
plt.title('Inverted Binary')
plt.imshow(inv_img, cmap = 'gray')
```

##### Otsu Thresholding

```python
Thresh_value, thresh_otsu = cv2.threshold(img,0,255,cv2.THRESH_OTSU)
```

```python
Thresh_value
```

```python
plt.figure(figsize=(16,9))
plt.subplot(141)
plt.title('original')
plt.imshow(img, cmap = 'gray')
plt.subplot(142)
plt.title('Binary')
plt.imshow(new_img, cmap = 'gray')
plt.subplot(143)
plt.title('Inverted Binary')
plt.imshow(inv_img, cmap = 'gray')
plt.subplot(144)
plt.title('Threshold')
plt.imshow(thresh_otsu, cmap = 'gray')
```
