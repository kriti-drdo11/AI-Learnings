```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('images/lenna.png')
```

```python
resized_image = cv2.resize(image,(100,200))

cv2.imshow('resized image',resized_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
resized_image.shape
```

```python
resized_image = cv2.resize(image,(700,800))

cv2.imshow('resized image',resized_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

##### scaling Factor

```python
resized_image = cv2.resize(image,None, fx = 0.75, fy= 0.75)

cv2.imshow('resized image',resized_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
resized_image.shape
```

```python
resized_image = cv2.resize(image,None, fx = 1.75, fy= 1.75)

cv2.imshow('resized image',resized_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
resized_image.shape
```

##### scikit Image

```python
from skimage.transform import resize
```

```python
image = plt.imread('images/lenna.png')
```

```python
resized_img = resize(image,(1000,2000),anti_aliasing=True)
plt.subplot(1,2,1)
plt.imshow(image)
plt.subplot(1,2,2)
plt.imshow(resized_img)
```
