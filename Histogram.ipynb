```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
num = [np.random.randint(0,100) for x in range (1000)]
num
```

```python
print(num)
```

```python
plt.title('Histogram')
plt.xlabel('values')
plt.ylabel('count')
plt.xticks(range(10,101,10))
plt.grid()
plt.hist(num,bins = range(10,101,10))
```

```python

img = cv2.imread('images/lenna.png',0)
plt.figure(figsize= (16,9))
plt.title('Image Histogram')
plt.xlabel('values')
plt.ylabel('count')
plt.xticks(range(0,256,10))
plt.grid()
plt.hist(img.flatten(),bins = range(0,256,10))
```
