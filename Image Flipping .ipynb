```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('images/lenna.png')
flip_h = cv2.flip(image,0)
flip_v = cv2.flip(image,1)
flip_d = cv2.flip(image,-1)

cv2.imshow('original',image)
cv2.imshow('Horizontal flip',flip_h)
cv2.imshow('vertical flip',flip_v)
cv2.imshow('Digonal flip',flip_d)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
flip_hor = cv2.cvtColor(flip_h,cv2.COLOR_BGR2RGB)
flip_verr = cv2.cvtColor(flip_v,cv2.COLOR_BGR2RGB)
flip_dig = cv2.cvtColor(flip_d,cv2.COLOR_BGR2RGB)
flip_org = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```
