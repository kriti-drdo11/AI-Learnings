```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread('images/eiphel1.png',0)
img2 = cv2.imread('images/eiphel2.png',0)
```

```python
plt.figure(figsize= (16,9))
plt.subplot(1,2,1)
plt.imshow(img1,cmap ='gray')
plt.subplot(1,2,2)
plt.imshow(img2,cmap ='gray')
```

```python
sift = cv2.SIFT_create()
```

```python
keypt1 ,desc1 = sift.detectAndCompute(img1,None)
keypt2 ,desc2 = sift.detectAndCompute(img2,None)
len(keypt1), len(keypt2)
```

```python
bf = cv2.BFMatcher(cv2.NORM_L1)

matchs = bf.match(desc1,desc2)
matchs = sorted(matchs,key = lambda x:x.distance)

img3 = cv2.drawMatches(img1,keypt1,img2,keypt2,matchs[:10],img2,flags = 2)
plt.imshow(img3)
```
